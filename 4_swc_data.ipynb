```python
from os import listdir
from os.path import isfile, join
import pandas as pd
from pydub import AudioSegment
import os
import shutil
import natsort 
import wave
import xml.etree.ElementTree as ET
#https://github.com/BayesForDays/swc_parser
```

```python
rootdir= "/home/megha/Desktop/b/data" 
#rootdir= "F:/Thesis/1/data_temp"
for Dir in natsort.natsorted(os.listdir(rootdir)):
    for file in os.listdir(rootdir+"/"+Dir):
        if(file.endswith(".swc")):
            tree = ET.parse(rootdir+"/"+Dir+"/"+file)
            root = tree.getroot()
            time_start_array = []
            time_end_array = []
            text_array = []
            for children in root.findall("./d"):
                for child in children:
                    print(child.tag)# get the child details here
                    if(child.tag == "extra"):
                        for n in child.iter('n'):
                            text_array.append(n.attrib['pronunciation'])
                            if('start' in (n.attrib)):
                                time_start_array.append(n.attrib['start'])
                            else:
                                time_start_array.append(0)
                            if('end' in (n.attrib)):
                                time_end_array.append(n.attrib['end'])
                            else:
                                time_end_array.append(0)
                        print(time_start_array)
                        print(time_end_array)
                        #print(text_array)
                    if(child.tag == "p"):
                        for n in child.iter('n'):
                            #print(n.attrib['pronunciation'])
                            #print(n.attrib['start'])
                            #print(n.attrib['end'])
                            text_array.append(n.attrib['pronunciation'])
                            if('start' in (n.attrib)):
                                time_start_array.append(n.attrib['start'])
                            else:
                                time_start_array.append(0)
                            if('end' in (n.attrib)):
                                time_end_array.append(n.attrib['end'])
                            else:
                                time_end_array.append(0)
                        print(time_start_array)
                        print(time_end_array)
                    if(child.tag =="section"):
                        for n in child.iter('n'):
                            #print(n.attrib['pronunciation'])
                            #print(n.attrib['start'])
                            #print(n.attrib['end'])
                            text_array.append(n.attrib['pronunciation'])
                            if('start' in (n.attrib)):
                                time_start_array.append(n.attrib['start'])
                            else:
                                time_start_array.append(0)
                            if('end' in (n.attrib)):
                                time_end_array.append(n.attrib['end'])
                            else:
                                time_end_array.append(0)
                        print(time_start_array)
                        print(time_end_array)
                    if(child.tag =="ignored"):
                        for n in child.iter('n'):
                            #print(n.attrib['pronunciation'])
                            #print(n.attrib['start'])
                            #print(n.attrib['end'])
                            text_array.append(n.attrib['pronunciation'])
                            if('start' in (n.attrib)):
                                time_start_array.append(n.attrib['start'])
                            else:
                                time_start_array.append(0)
                            if('end' in (n.attrib)):
                                time_end_array.append(n.attrib['end'])
                            else:
                                time_end_array.append(0) 
                        print(time_start_array)
                        print(time_end_array)
                df = pd.DataFrame({'col1':time_start_array, 'col2':time_end_array, 'col3':text_array})        
                #for child1 in child.iter("t"):
                 #   print(child.attrib)
                #print(dee.child)
                #children = dee.getchildren()
                #print(children)
                #for child in children:
                #    print(child.tag)
                #    if(child.tag == "extra"):
                #        for c in child:
                #            print('hi')
                #            #if(c.)
```
