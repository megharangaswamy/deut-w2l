```python
import webvtt
from os import listdir
from os.path import isfile, join
import pandas as pd
from pydub import AudioSegment
import os
import shutil
import natsort 
import sys, nltk
import codecs
```

```python
# I have keept only german .txt files from all the 9 folders to make the code simple.
rootdir = "/home/megha/Desktop/d"
```

```python
punctuations = '''!,.?abcdefghijklmnopqrstuvwxyzäüößABCDEFGHIJKLMNOPQRSTUVWXYZÄÜÖẞ '''
between_digits = ''',.0123456789'''

with open("/home/megha/Desktop/LM/Kaggle.txt", "a") as write_file:
    for file in natsort.natsorted(os.listdir(rootdir)):
        print(file)
        print('---------------------------------------------------------------')
        if(file[-4:] == ".txt"):
            #with codecs.open(rootdir+"/"+file, 'r', encoding='utf8') as f:
            with open(rootdir+"/"+file, 'rb') as f:
                line = f.readline()
                cnt = 0
                while line:
                    no_punct = ""
                    i = 0
                    while i < len(line):
                        if(i != (len(line)-1)):
                            if((str(line[i]).isdigit() and line[i+1] in between_digits) or (line[i] in between_digits and str(line[i+1]).isdigit())):
                                i = i+1
                            else:
                                if line[i] in punctuations:
                                    no_punct = no_punct + line[i]
                        else:
                            if line[i] in punctuations:
                                no_punct = no_punct + line[i]
                        i += 1
                    for sentence in nltk.sent_tokenize(no_punct):
                        line_mid = ' '.join(nltk.word_tokenize(sentence)).lower()
                        if(line_mid != "."):
                            write_file.write(line_mid+'\n')  
                            #print(line_mid)
                            #print('------------------------------------')
                    line = f.readline()        
                    cnt += 1
```
